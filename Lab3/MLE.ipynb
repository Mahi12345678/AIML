```python
import numpy as np

np.random.seed(1)

true_p=0.6
data=np.random.binomial(1, true_p,10)

print("Observed coin tosses (1=Head, 0=Tail):")
print(data)
print("Number of heads:", data.sum())
```

```python
def log_likelihood(p, data):
    return np.sum(data*np.log(p)+(1-data)*np.log(1-p))
```

```python
import matplotlib.pyplot as plt
p_values=np.linspace(0.01,0.99,100)
ll_values=[log_likelihood(p,data) for p in p_values]
```

```python
plt.plot(p_values, ll_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

true_p=0.7
sample_size=[10,30,50,100,500,1000]
p_mle_values=[]

for n in sample_size:
    data=np.random.binomial(1, true_p, n)
    p_mle=data.mean()
    p_mle_values.append(p_mle)

plt.plot(sample_size, p_mle_values, marker='o')
plt.axhline(true_p, linestyle='--')
plt.xlabel('Sample Size')
plt.ylabel('MLE of p')
plt.title('Convergence of MLE to True Parameter Value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# True parameters
true_mean = 5
true_sd = 2

sample_size = [10, 30, 50, 100, 500, 1000]
mean_mle_values = []

for n in sample_size:
    data = np.random.normal(true_mean, true_sd, n)
    mean_mle = data.mean()   # MLE of mean
    mean_mle_values.append(mean_mle)

plt.plot(sample_size, mean_mle_values, marker='o', label='MLE of mean')
plt.axhline(true_mean, linestyle='--', label='True mean')
plt.xlabel('Sample Size')
plt.ylabel('MLE of Mean')
plt.title('Convergence of MLE to True Mean')
plt.legend()
plt.show()
```
